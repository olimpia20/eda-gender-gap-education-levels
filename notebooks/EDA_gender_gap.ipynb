Clean Data

```python
# Define the columns to rename
rename_map = {
    'Combined gross enrolment ratio for tertiary education, female' : 'ter_female',
    'Combined gross enrolment ratio for tertiary education, male' : 'ter_male',
    'Combined total net enrolment rate, secondary, male': 'sec_male',
    'Combined total net enrolment rate, secondary, female': 'sec_female'
}

# Rename the columns first
df_renamed = df.rename(columns=rename_map)

# 3. Select ONLY the renamed columns AND any other columns you need (e.g., 'Entity', 'Year')
# You must include the original columns you want to keep, like 'Entity' and 'Year'.
# I'll assume you want 'Entity' and 'Year' based on common dataset structure.

desired_columns = ['Entity', 'Year', 'ter_female', 'ter_male', 'sec_male', 'sec_female'] 

df_clean = df_renamed[desired_columns]
```

```python
print("Cleaned dataset shape:", df_clean.shape)
print("\nFirst few rows:")
print(df_clean.head())
print("\nColumn names:")
print(df_clean.columns.tolist())
```

Keep only countries from Europe

```python
# The list of European countries
europe_countries = [
    'Albania', 'Andorra', 'Austria', 'Belarus', 'Belgium', 'Bosnia and Herzegovina',
    'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 'Denmark', 'Estonia', 'Finland', 
    'France', 'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland', 'Italy', 
    'Kosovo', 'Latvia', 'Liechtenstein', 'Lithuania', 'Luxembourg', 'Malta', 
    'Moldova', 'Monaco', 'Montenegro', 'Netherlands', 'North Macedonia', 'Norway', 
    'Poland', 'Portugal', 'Romania', 'Russia', 'San Marino', 'Serbia', 'Slovakia', 
    'Slovenia', 'Spain', 'Sweden', 'Switzerland', 'Ukraine', 'United Kingdom', 'Vatican' 
]

# Create a boolean mask and filter the DataFrame
df_europe = df_clean[df_clean['Entity'].isin(europe_countries)]
```
